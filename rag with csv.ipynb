```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

#embeddings = OpenAIEmbeddings()
# A common embedding model
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

index = faiss.IndexFlatL2(len(HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
).embed_query(" ")))
vector_store = FAISS(
    embedding_function=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
),
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
```

```python
vector_store.add_documents(documents=docs)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel, RunnablePassthrough

retriever = vector_store.as_retriever()

system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you "
    "don't know. Use three sentences maximum and keep the "
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

# This replaces create_stuff_documents_chain(llm, prompt)
question_answer_chain = prompt | llm | StrOutputParser()

# This replaces create_retrieval_chain(retriever, question_answer_chain)
rag_chain = (
    RunnableParallel(
        context=lambda x: retriever.invoke(x["input"]),
        input=RunnablePassthrough()
    )
    | question_answer_chain
)

# Example call:
result = rag_chain.invoke({"input": "According to the \"Resetting a Forgotten PIN\" section, what is the first step an employee should take to begin the process of resetting their PIN?"})
print(result)
```

1. Loading CSV Data
python
from langchain_community.document_loaders.csv_loader import CSVLoader
loader = CSVLoader(file_path=file_path)
docs = loader.load_and_split()
Purpose: Loads data from a CSV file and splits it into smaller chunks (documents) suitable for processing.

How it works: Each row (or group of rows) becomes a Document object with a page_content field containing the text.

2. Setting Up FAISS Vector Store
python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
index = faiss.IndexFlatL2(len(embeddings.embed_query(" ")))
vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
Purpose: Creates a local vector database (FAISS) to store and search document embeddings.

Key steps:

HuggingFaceEmbeddings: Generates numerical embeddings for text using the specified model.

faiss.IndexFlatL2: Initializes a FAISS index for storing vectors and performing similarity searches.

FAISS(...): Wraps the FAISS index, embedding function, and document store into a LangChain vector store.

3. Creating a Retriever
python
retriever = vector_store.as_retriever()
Purpose: Enables similarity search on the vector store. When given a query, it retrieves the most relevant documents.

4. Setting Up Prompt and LLM Chain
python
system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you "
    "don't know. Use three sentences maximum and keep the "
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

question_answer_chain = prompt | llm | StrOutputParser()
Purpose: Defines how to format the question and context for the LLM and how to parse its response.

How it works:

ChatPromptTemplate: Combines the system prompt and user input into a message sequence.

question_answer_chain: A chain that formats the prompt, sends it to the LLM, and parses the output into a string.

5. Building the RAG Chain
python
rag_chain = (
    RunnableParallel(
        context=lambda x: retriever.invoke(x["input"]),
        input=RunnablePassthrough()
    )
    | question_answer_chain
)
Purpose: Combines retrieval and question-answering into a single chain.

How it works:

RunnableParallel: Retrieves relevant context and passes the original input through.

| question_answer_chain: Feeds the retrieved context and input into the LLM chain.

6. Running the Chain
python
result = rag_chain.invoke({"input": "which company does sheryl Baxter work for?"})
print(result)
Purpose: Executes the RAG pipeline for a specific question.

How it works:

The input question is passed to the retriever to find relevant documents.

The retrieved context and the original question are sent to the LLM.

The LLM generates an answer based on the context.

Summary Table
Step	Purpose	Key Components Used
CSV Loading	Load and split CSV data	CSVLoader
Vector Store Setup	Store document embeddings	FAISS, HuggingFaceEmbeddings, faiss
Retriever	Retrieve relevant documents	as_retriever()
Prompt and Chain	Format and parse LLM input/output	ChatPromptTemplate, StrOutputParser
RAG Chain	Combine retrieval and QA	RunnableParallel, RunnablePassthrough
Running the Chain	Answer a question using RAG	invoke()
